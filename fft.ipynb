```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier CSV contenant le signal ECG
data = pd.read_csv('donnees_filtrees.csv')

# Extraire la colonne du signal ECG
ecg_signal = data['MLII_filtered'].values
ecg = pd.DataFrame(np.array(ecg_signal).reshape(-1, 1))

# Calculer le spectre de frquences
n = len(ecg)
timestep = 1 / 360  # Priode d'chantillonnage
freq = np.fft.fftfreq(n, timestep)
ecg_spectrum = np.fft.fft(ecg)

# Tracer le spectre de frquences
plt.figure(figsize=(10, 6))
plt.plot(freq[:n//2], np.abs(ecg_spectrum)[:n//2])
plt.xlabel('Frquence (Hz)')
plt.ylabel('Amplitude')
plt.title('Spectre frquentiel du signal ECG')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier CSV contenant le signal ECG
data = pd.read_csv('donnees_filtrees.csv', sep=',')

# Extraire la colonne du signal ECG
ecg_signal = data['MLII_filtered'].values

# Calculer le spectre de frquences
n = len(ecg_signal)
timestep = 1 / 360  # Priode d'chantillonnage
freq = np.fft.fftfreq(n, timestep)
ecg_spectrum = np.fft.fft(ecg_signal)

# Trouver l'indice de la plus haute frquence
index_max_freq = np.argmax(np.abs(ecg_spectrum))

# Obtenir la plus haute frquence et son amplitude
max_freq = freq[index_max_freq]
max_amplitude = np.abs(ecg_spectrum)[index_max_freq]

# Trouver la plus grande frquence
max_frequency_value = np.max(freq)

# Afficher la valeur de la plus haute frquence et de la plus grande frquence
print(f"Plus haute frquence: {max_freq:.2f} Hz")
print(f"Plus grande frquence: {max_frequency_value:.2f} Hz")

# Tracer le spectre de frquences
plt.figure(figsize=(10, 6))
plt.plot(freq[:n//2], np.abs(ecg_spectrum)[:n//2], color='blue')
plt.xlabel('Frquence (Hz)')
plt.ylabel('Amplitude')
plt.title('Spectre frquentiel du signal ECG')
plt.grid(True)

# Marquer la plus haute frquence avec un point rouge
plt.scatter(max_freq, max_amplitude, color='red', label=f'Frquence max = {max_freq:.2f} Hz')

# Tracer la raie en rouge
plt.vlines(max_freq, 0, max_amplitude, colors='red', linestyles='dashed', label='Raie')

plt.legend()
plt.show()


```
